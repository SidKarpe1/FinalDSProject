Now we’ll create two functions, one that calculates the quantity of times a value shows up in a column and one that creates a bar graph. We’re putting this code inside functions because we’ll need this code for each column and it’ll be easier to call a function multiple times rather than code the same graph and dictionary every time. Now we’ll create two functions, one that calculates the quantity of times a value shows up in a column and one that creates a bar graph. We’re putting this code inside functions because we’ll need this code for each column and it’ll be easier to call a function multiple times rather than code the same graph and dictionary every time.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('ds_salaries.csv')
# function that calculates the quantity of values in a table
def frequency(list1):
  dict1 = {}
  for i in list1.to_numpy():
        if i not in dict1:
          dict1[i] = 1
        else:
          dict1[i] += 1
  return dict1

def bar_graph(data): # creating a function that makes a bar graph for any column of data
    dict1 = frequency(df[data])
    categories = dict1.keys() # categories as list
    values = dict1.values() # frequencies as lists
    plt.bar(categories, values, color='skyblue')
    plt.xticks(rotation=90)
    plt.xlabel(data)
    plt.ylabel('Frequency')
    plt.title('Bar Graph')
    plt.show()

#calling the bar graph functions
bar_graph('experience_level')
bar_graph('employment_type')
bar_graph('job_title')
```

##Insights

Looking at these graphs and queries, we can see that the data is quite unevenly distributed. This means that analysis will be better for certain categories rather than others, however, it is still possible because we have a wide variety of data.

Looking at the array above, we see that there are four levels of experience, EN, MI, SE and EX. Every employee in the dataset is assigned one of these levels and a salary. To analyze how the salary changes, we first need to create a table that has each of these levels as their own column and have all the matching salaries as cells in this table. To do this, we need to create arrays with all the salaries and assign them to a key in a dictionary, however, instead of coding it separately for every rank, we’ll can do all of them at once with a for loop.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('ds_salaries.csv')

# Query: Find unique values in experience level
unique = df['experience_level'].unique()
# Query: Experience level uniques
entries = df['experience_level']

dict1 = {}
for i in unique:
  # Query all values with either EN, MI... based on iteration
  rows = df.loc[entries.str.contains(i, na=False)]
  # Store the salary in USD values
  values = rows['salary_in_usd'].to_numpy()
  # Create dictionary with EN,MI... as Key and respective salaries as values
  dict1[i] = values
#Query: reorder in ascending value of rank
temp = ['EN', 'MI', 'SE', 'EX']
salDict = {key: dict1[key] for key in temp}
# Fill in empty values with Null
salByRank = pd.DataFrame({key: pd.Series(value) for key, value in salDict.items()})
#Query: return count, mean, median, quartiles, etc.
salByRank.describe()
```

#Question 3: What is the best position in data science in terms of salary

To find the best job position, we’ll look at the entry job positions to see which job you should try to start out as. In order to do this, we need to take a few steps. Our first step is to locate all beginner level employees. Our second step is to make a table with these employees’ job titles and salaries. Since this table has multiple people with the same job title, we’ll group people will the same position into one column and average their salaries. Our last step is to reorder the table by descending salary to find the highest paid beginner positions.


```python
import pandas as pd
import numpy as np
df = pd.read_csv('ds_salaries.csv')
#query: locating beginner entries
beg = df.loc[df['experience_level'] == 'EN']
#making a dataframe with positions and salaries
tempDict = {'job': beg['job_title'].to_numpy(), 'salary in USD': beg['salary_in_usd']}
newdf = pd.DataFrame(tempDict)

#query: grouping the values
grouped = newdf.groupby(newdf['job'])['salary in USD'].mean().to_numpy()
jobs = np.sort(beg['job_title'].unique()) # query:  find unique values in job title column in alphabetical order
#query: reordering the data
tempDict['job'] = jobs
tempDict['salary in USD'] = grouped
newdf = pd.DataFrame(tempDict).sort_values(by="salary in USD", ascending = False)
newdf.head(5)
```

##Insights

After printing the top 5 jobs, we can see that our best options as a beginner are Machine Learning Scientist, Applied Data Scientist, Computer Vision Software Engineer, Financial Data Analyst and Machine Learning Developer. Seeing as how all of these jobs average around or above $100k, we can answer our question saying these are the best positions to apply for when looking for a job in data science.

#Question 5: Should you live where your company is based

Answering this question will be a little bit tricky so we’ll take it step by step. Our first step is to create two dataframes. Our first dataframe will be all the entries where people live where their companies are based and their second dataframe will be all the entries where people live in a different country than where their company is based. To do this, we’ll use conditionals finding where the two columns match and where the two columns dont match. After this, we’ll find median salary in USD and compare the two values.


```python
import pandas as pd
import numpy as np
df = pd.read_csv('ds_salaries.csv')

# querying where people live where they work
sameCountry = df[df['employee_residence'] == df['company_location']]
# querying where people live in a different country than where they work
diffCountry = df[df['employee_residence'] != df['company_location']]
# query: finding the median of both columns
sameMedian = sameCountry['salary_in_usd'].median()
diffMedian = diffCountry['salary_in_usd'].median()
print(f' people who live in the country where they work typically make around {int(sameMedian)}, \n people who live in a different country than where they work typically make around {int(diffMedian)}, \n people who live where they work typically make around {int(sameMedian - diffMedian)} more than those who do not.')
```

#Insights

Looking at the two statements, we see that people who live in the same country where they work tend to make a lot more money with a different of over 40k! This means that if you want to make good money while working in Data Science its best to either find a company where you live or get ready to move where you get hired.


#Question 6: How does working remotely affect your salary

```python
import pandas as pd
import numpy as np
df = pd.read_csv('ds_salaries.csv')

#query: find unique remote values and the data type of each element
remote = df['remote_ratio'].unique()
print(remote, type(remote[0]))
```

Here we see that there are three elements, 0, 50 and 100. We also see that all of these values are integers. In order to make the analysis easier, it will probably be better to use strings, so we’ll covert these values in strings and reimport them into our table before working beginning our analysis .


```python
import pandas as pd
import numpy as np
df = pd.read_csv('ds_salaries.csv')
remote = df['remote_ratio'].to_numpy()

#turn values into strings and put them back into dataframe
remote = np.array([str(i) for i in remote])
df['remote_ratio'] = remote

#step 1
office = df.loc[df['remote_ratio']=='0']
half = df.loc[df['remote_ratio']=='50']
home = df.loc[df['remote_ratio']=='100']
#step 2
office = office['salary_in_usd'].to_numpy()
half = half['salary_in_usd'].to_numpy()
home = home['salary_in_usd'].to_numpy()
#step 3
office = np.median(office)
half = np.median(half)
home = np.median(home)
print(f' The median income of people who do not work remotely is is {office}. \n The median income of people who work at home {home}. \n The median income of people who work at home but also go to the office is {half}.')
```

##Insights

Now that we can see our median values, we can conclude that working at home is more efficient and has a higher income than at the office or doing both.


#Question 7: Has the average salary increased over time



For this question, we’re going to create a graph since it will be easier to vizualize how salaries increase. In order to do this, we’re first going to create two columns, one with the years as a string and the other with the mean salary of each year as an integer. We’ll then use a for loop to create an array with all the salaries for each year and store them in a dictionary as a value assigned to a year as a key. Finally, we can use matplotlib to create a line graph to show us how the salaries change over time.


```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('ds_salaries.csv')
#query: find unique years and turning them into strings
years = df['work_year'].unique()
years = [str(i) for i in years]
values = df['work_year']
#creating a dataframe with years as columns and salaries as entries
tempdict = {}
for i in years:
  tempvar = df.loc[df['work_year']==i]['salary_in_usd'].to_numpy()
  tempdict[i] = tempvar
# query: group by position with mean salary
sal = df.groupby('work_year')['salary_in_usd'].mean().to_numpy()
# creating a plot with years as x value and salaries as y values
plt.plot(years, sal)
plt.xticks(rotation=45)
plt.xlabel('years')
plt.ylabel('salary')
plt.title('average salary over the years')
plt.show()
```

Now, we'll create a dataframe with a column that shows the rate of change over the years. To do this, we start by putting these columns in a dictionary and turning it into a dataframe. Then, we can use the pandas function .pct_change() to calculate the rate of change.

```python
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('ds_salaries.csv')

years = df['work_year'].unique()
years = [str(i) for i in years]
sal = df.groupby('work_year')['salary_in_usd'].mean().to_numpy()
tempDict = {'years': years,
            'salary': sal}
df = pd.DataFrame(tempDict)
# query finding the difference in rows.
df['rate_of_change'] = df['salary'].pct_change() * 100
df
```

##Insight

Looking at the graph and the table, we can see that there is a more exponential growth curve to the data rather than a linear one. Using this information, we can answer our question and claim that salaries increase over the years due to demand and that data science is a good career path to pursue since your salary is likely to increase over the years along with other factors such as experience level and change in higher level positions such as manager and director.
